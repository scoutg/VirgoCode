Filtering data by mass constraints, log(Mstar/Msol)>8,9,10,11

```python
constraint = 8
length = data.field('MSTAR_50').size
mass = data.field('MSTAR_50')
fRA = [0]*length
fDEC = [0]*length
```

```python
length
```

This intitial method of filtering was not successful because it creates many 0 valued points

```python
for i in range (0, length):
    if mass[i] > constraint:
        fRA[i] = RA[i]
        fDEC[i] = DEC[i]
```

```python
massb = mass[mass>constraint]
lengthb = massb.size
fRAb = [0]*lengthb
fDECb = [0]*lengthb
fZb = [0]*lengthb
a = 0
```

```python
for i in range (0, length):
    if mass[i] > constraint:
        fRAb[a] = RA[i]
        fDECb[a] = DEC[i]
        fZb[a] = data.field('Z')[i]
        a = a + 1
```

The first plot is without the color-coding.

```python
plt.plot(fRAb, fDECb, 'ko', ms = 3, label = 'Filtered Data')
plt.gca().invert_yaxis()
plt.show()
```

```python
The next plot and all of the ones following it are color-coded based on z.
```

```python
plt.scatter(fRAb, fDECb, s= 8, c=fZb, cmap='magma')
plt.colorbar()
plt.gca().invert_yaxis()
plt.show()
```
