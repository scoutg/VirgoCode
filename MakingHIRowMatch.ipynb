Need to open for dimensions and galaxy list: CO-MasterFile-2018Feb16.fits

```python
masterFile = fits.open('CO-MasterFile-2018Feb16.fits')
```

```python
masterfileData = masterFile[1].data
masterCols = masterFile[1].columns
masterCols
```

```python
NEDnames = masterfileData.field('NED_name')
SDSSID = masterfileData.field('SDSS_ID')
```

NED names are the ones that will match names in all-HI.2018Feb18.txt

```python
Length = NEDnames.size
Length
```

```python
data = pd.read_csv('all-HI.2018Feb18.tc.txt', delim_whitespace = True)
data
```

```python
Area = [0]*Length
A_err = [0]*Length
V = [0]*Length
V_err = [0]*Length
FWHM = [0]*Length
fwhm_err = [0]*Length
Tpeak = [0]*Length
Tpeak_err = [0]*Length
```

```python
for i in range (0, Length):
    for j in range (0, 64):
        if NEDnames[i] == data['Source'][j]:
            Area[i] = data['Area'][j]
            A_err[i] = data['err'][j]
            V[i] = data['V'][j]
            V_err[i] = data['err.1'][j]
            FWHM[i] = data['FWHM'][j]
            fwhm_err[i] = data['err.2'][j]
            Tpeak[i] = data['Tpeak'][j]
            Tpeak_err[i] = data['err.3'][j]
      
```

```python
col1 = fits.Column(name='Names', format='20A', array=NEDnames)
col2 = fits.Column(name='Area', format='20A', array=Area)
col3 = fits.Column(name='Err_A', format='20A', array=A_err)
col4 = fits.Column(name='V', format='20A', array=V)
col5 = fits.Column(name='Err_V', format='20A', array=V_err)
col6 = fits.Column(name='FWHM', format='20A', array=FWHM)
col7 = fits.Column(name='Err_FWHM', format='20A', array=fwhm_err)
col8 = fits.Column(name='Tpeak', format='20A', array=Tpeak)
col9 = fits.Column(name='Err_Tpeak', format='20A', array=Tpeak_err)
```

```python
cols = fits.ColDefs([col1, col2, col3, col4, col5, col6, col7, col8, col9])
```

```python
hdu = fits.BinTableHDU.from_columns(cols)
```

```python
hdu.writeto('HIRowMatch.fits')
```
